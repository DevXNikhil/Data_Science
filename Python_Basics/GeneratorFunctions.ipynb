Generator Functions - A generator function is a function that gives you values one by one instead of all at once.

```python
# Example of range function 
for i in range(1,11): # It will give data one by one from 10 to 100
    print(i)
```

```python
# Example - 1 Fibonacci series using while loop
def fibconacci(n):
    a,b = 0,1
    i = 0
    while i < n:
        print(a)
        c = a+b;
        a = b
        b = c
        i +=1 
```

```python
fibconacci(10)
```

```python
# Example - 2 Fibonacci series using generator function with for loop
def fibo_genFunc(n):
    a,b = 0,1
    for i in range(n):
        yield a # use yield keywd instead of return 
        a,b = b , a+b
```

```python
for i in fibo_genFunc(10):
    print(i)
```

```python
 # Example - 3 Fibonacci series using generator function with while loop
def fibo_genFunc():
    a,b = 0,1
    while True:
        yield a
        a,b = b , a+b
```

```python
fib = fibo_genFunc() # Object Creation
```

```python
type(fib)
```

```python
for i in range(10):
    print(next(fib))
```

```python
# Example - 4 
def count_test(n):
    count = 1
    while count <= n :
        yield count
        count += 1
```

```python
c = count_test(5) # Object 
```

```python
for i in c:
    print(i)
```
