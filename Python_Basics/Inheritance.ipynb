# Inheritance - in OOP is a concept where a child class acquires the properties and methods of a parent class.

```python
# Example - 1 Single Inheritance
class parent: # class 1
    def method(self): # self -> pointer 
        return "Parent class"
```

```python
class child(parent): # class 2
    def function(self):
        return "Child class"
```

```python
child_obj = child()  # Object creation of child class
```

```python
child_obj.method() # Calling method of parent class using object of child class
```

```python
child_obj.function()
```

```python
# Example - 2 Multilevel Inheritance
class class1(): # Class 1

    def meth(self):
        return "This is method from class1"
```

```python
class class2(class1): # Class 2

    def method(self):
        return "This is method from class2"
```

```python
class class3(class2): # Class 3

    def func(self):
        return "This is method from class3"
```

```python
obj_class3 = class3() # Object creation for class3 only
```

```python
obj_class3.method() 
```

```python
obj_class3.meth()
```

```python
obj_class3.func()
```

```python
# Example - 3  Multiple Inheritance - Single class can inherit properties of 2 parent classes
class cls1:
    def meth(self):
        return "This is class 1"
```

```python
class cls2:
    def method(self):
        return "This is class 2"
```

```python
class cls3(cls1,cls2):
    def function(self):
        return "This is class 3"
```

```python
obj_class3 = cls3() # Object creation of class 3
```

```python
obj_class3.method() # accessing method of cls2
```

```python
obj_class3.meth()  # accessing method of cls1
```

```python
obj_class3.function() # accessing method of its own
```
