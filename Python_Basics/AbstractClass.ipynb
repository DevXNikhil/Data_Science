```python
# Abstract Class
```

```python
# Example - 1
import abc    
class pwskills:  # Abstract class
    @abc.abstractmethod
    def student_detail(self):  # Methods are only declared - Prototype
        pass

    @abc.abstractmethod
    def student_assign(self):
        pass

    @abc.abstractmethod
    def student_marks(self):
        pass
```

```python
class data_science(pwskills): # Inheriting properties of pwskill class
    
    def student_detail(self): # Methods are defined - definition of function
        return "Students of data science"

    def student_assign(self):
        return "Assignments of data science"

    def student_marks(self):
        return "Students marks of data science"
```

```python
class web_development(pwskills): # Inheriting properties of pwskill class

    def student_detail(self): # Methods are defined - definition of function
        return "Students of web development "

    def student_assign(self):
        return "Assignment of web development "

    def student_marks(self):
        return "Students marks of web development"
```

```python
ds = data_science() # Object of data science class
```

```python
ds.student_detail() # Accessing methods of data science class
```

```python
ds.student_assign()
```

```python
ds.student_marks()
```

```python
wd = web_development() # Object of web development class
```

```python
wd.student_detail() # Accessing methods of web development class
```

```python
wd.student_assign()
```

```python
wd.student_marks()
```
