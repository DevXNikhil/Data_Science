# Decorators 

```python
# Example : 1
def test():
    print("This is the start of function") # To avoid doing the same work repeatedly, we use decorators.
    print(4+6)
    print(10/2)
    print("This is the end of function")
```

```python
test()
```

```python
# Example - 2
# decorator function for repeating one task
# So we can do one task multiple time using decorator
def decorator(function):
    def inner_deco():
        print("Starting")
        function()
        print("Ending")
    return inner_deco
```

```python
@decorator # we have to write decorator name like - @decorator in the function block
def add():
    print(5+5)
```

```python
add()
```

```python
# Function - 2 here without writing starting and ending statement we can see both statements because of decorator
@decorator
def multi():
    print(5*5)
```

```python
multi()
```

```python
 # Example - 3 Decorators are used for checking the running time of a function.
import time   # for time package

def time_test(function):   # Accepting function as input
    def inner_time_test():
        start = time.time()  # calling time function
        function()  # calling function
        end = time.time()
        print(end - start)  # It prints the time taken to run a function.
    return inner_time_test 
```

```python
@time_test
def addition(): # Function 1
    print(50+50)
```

```python
addition()
```

```python
@time_test
def division(): # Function 2
    print(50/5)
```

```python
division()
```

```python
@time_test
def test(): # Function - 3
    for i in range(10000000):
        pass
```

```python
test()
```
